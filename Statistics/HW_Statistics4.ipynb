### Значения для столбцов

```python
def described(horses):
    for b in horses:
        print('mean ', b, pd.to_numeric(horses[b], errors='coerce').mean())
        print('median', b, pd.to_numeric(horses[b], errors='coerce').median())
        print('mode', b, pd.to_numeric(horses[b], errors='coerce').mode()[0])
        print('std', b, pd.to_numeric(horses[b], errors='coerce').std())
        print('var', b, pd.to_numeric(horses[b], errors='coerce').var())
        print('max ', b, pd.to_numeric(horses[b], errors='coerce').max())
        print('minimum ', b, pd.to_numeric(horses[b], errors='coerce').min())
    
```

```python
described(sorted_horse_data)
```

###### Считаем пропуски в столбцах

```python
for x in sorted_horse_data:
    print('количество пропусков в - ', x, ' = ',  sorted_horse_data[sorted_horse_data[x] == '?'].count().unique())
    
```

```python
calc_values_mean = []
calc_values_median = []
calc_values_mode = []
calc_values_std = []
calc_values_var = []
calc_values_max = []
calc_values_min = []
for b in sorted_horse_data:
    calc_values_mean.append(pd.to_numeric(sorted_horse_data[b], errors='coerce').mean())
    calc_values_median.append(pd.to_numeric(sorted_horse_data[b], errors='coerce').median())
    calc_values_mode.append(pd.to_numeric(sorted_horse_data[b], errors='coerce').mode()[0])
    calc_values_std.append(pd.to_numeric(sorted_horse_data[b], errors='coerce').std())
    calc_values_var.append(pd.to_numeric(sorted_horse_data[b], errors='coerce').var())
    calc_values_max.append(pd.to_numeric(sorted_horse_data[b], errors='coerce').max())
    calc_values_min.append(pd.to_numeric(sorted_horse_data[b], errors='coerce').min())
```

```python
df = pd.DataFrame(list(zip(calc_values_mean, calc_values_median,calc_values_mode,calc_values_std,calc_values_var,calc_values_max
                          ,calc_values_min)),index=['surgery?', 'age', 'rectal temperature', 'pulse', 'respiratory rate',
                                                    'temperature of extremities', 'pain','outcome'],
                  columns=['mean', 'median','mode', 'std', 'var', 'Maximum', 'Minimum'])
```
