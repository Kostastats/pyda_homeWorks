Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies
```

```python
ratings = pd.read_csv('ratings.csv')
ratings
```

```python
joined = ratings.merge(movies, how='left', on='movieId')
joined
```

```python
def ratings_sorted(by_value):
    by_value.loc[(by_value.rating <= 2), 'class'] = 'низкий рейтинг'
    by_value.loc[(by_value.rating <= 4) & (by_value.rating > 2), 'class'] = 'средний рейтинг'
    by_value.loc[(by_value.rating > 4), 'class'] = 'высокий рейтинг'
    return by_value
```

```python
filmi = pd.read_csv('movies.csv')
otsenki = pd.read_csv('ratings.csv')
filmi
```

```python
otsenki
```

```python
years_out = filmi['title']
filmi[['titles', 'year']] = years_out.str.split('(', n=1, expand=True)
filmi.drop('title',axis =1, inplace=True) 
filmi
```

```python
filmi.year.str.rstrip(')')
```

```python
filmi_otsenki = filmi.merge(otsenki, how='left', on='movieId')
filmi_otsenki
```

```python
filmi_otsenki.replace({'year':'None'}, {'year': '1900'}, regex=True)
# ПЫТАЮСЬ ЗДЕСЬ ПОМЕНЯТЬ None НА 1900, ПОЧЕМУ-ТО НЕ УБИРАЕТСЯ.
```

```python
filmi_otsenki_year_filled = filmi_otsenki.fillna(value = {'year':'1900'})
```

```python
filmi_otsenki_year_filled.sort_values(by =[ 'rating'], ascending=False)
# ОТСОРТИРОВАЛ ПО РЕЙТИНГУ
```
