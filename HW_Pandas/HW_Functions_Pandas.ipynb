```python
movies = pd.read_csv('movies.csv')
movies
```

```python
ratings = pd.read_csv('ratings.csv')
ratings
```

```python
joined = ratings.merge(movies, how='left', on='movieId')
joined
```

```python
joined.loc[(joined.rating <= 2), 'class'] = 'низкий рейтинг'
joined.loc[(joined.rating <= 4) & (joined.rating > 2), 'class'] = 'средний рейтинг'
joined.loc[(joined.rating > 4), 'class'] = 'высокий рейтинг'
joined.head(40)
```

```python
def class_movie():
    if joined.loc[(joined.rating <= 2)]:
        joined['class'] = 'низкий рейтинг'
    elif joined.loc[(joined.rating <= 4) & (joined.rating > 2)]:
        joine['class'] = 'средний рейтинг'
    else:
        joined['class'] = 'высокий рейтинг'
    return class_movie
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}
print(geo_data['Центр'])
```

```python
geo_data = pd.DataFrame(geo_data)
geo_data
```

```python
geo_data = geo_data.rename(columns = {'Центр': 'Center', 'Северо-Запад': 'N-W', 'Дальний Восток': 'Far_East'})
geo_data
```

```python
regions.loc[(regions.keyword == geo_data.Center), 'region'] = 'Центр'
```

```python
 def sorted_regions():
    for sorted_ in regions:
        if regions[regions.keyword.str.contains(geo_data.Center)]:
            sorted_['region'] = 'Центр'
        elif regions[regions.keyword.str.contains(geo_data.N-W)]:
            sorted_['region'] = 'Северо Запад'
        elif regions[regions.keyword.str.contains(geo_data.Far_East)]:
            sorted_['region'] = 'Дальний Восток'
        else:
            sorted_['region'] = 'undefined'
        return regions
    
```
